**Find out the null value and if null value is found in any column then fill it with mean of that respective column**

**first check null value in each column**

```python
dt.isnull().sum()
```

**for filling the value lets suppose we take length column and lets fill the mean**

```python
dt['Length'].fillna(dt['Length'].mean())
```

**NOW CHECKING THE DIFFERENT TYPES OF THE MAKE ENTRY AND THE COUNTS OF EACH TYPE WHICH CAN BE DONE BY THE VALUE_COUNTS()**

```python
dt['Make'].value_counts()
```

**NOW GETTING ALL THE INSTANCES WHERE THE ORIGIN IS ASIA OR EUROPE**

```python
dt[(dt['Origin']=='Asia')|(dt['Origin']=='Europe')]
```

**HERE ONE THING TO NOTE THAT IF WE WANT TO REMOVE THE CERTAIN ROWS WITH A CONDITON THEN WHAT WILL DO IS 
dt[~condition] as we have done below**

```python
dt[~(dt['Weight']>4000)]
```
